# Simulation of a Sampling Distribution

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import default_rng
```

```python
rng = default_rng(32)

beta_0 = 7
beta_1 = 1.5
```

```python
x = np.array(list(np.linspace(start = 0, stop = 10, num = 11)) * 3)
n = 33
y = beta_0 + (beta_1 * x) + rng.standard_normal(n)
```

```python
plt.scatter(x =x, y = y)
plt.show()
```

```python
# module import
from sklearn.linear_model import LinearRegression

# list to store estimated slopes
slopes = []

# 5,000 simulations
for _ in range(5000):
  y = beta_0 + (beta_1 * x) + rng.standard_normal(33)

  # fit a linear regression model
  model = LinearRegression()
  model.fit(x.reshape(-1, 1), y)

  # store the estimated slope
  slopes.append(model.coef_[0])
```

```python
# plot a histrogram
plt.hist(slopes, bins = 20)
plt.xlabel('Estimated Slope')
plt.ylabel('# of Occurrences')
plt.title('Sampling Distribution of Estimated Slope')
plt.show()
```

```python
# compute the probability of getting an estimated slope greater than 1.65
probability = (np.sum(np.array(slopes) > 1.65)) / 5000
print(probability)
```

The calculated p-value in this study is 0.0038 (0.38%). This result indicates that, under the null hypothesis, the prorbability of observing a samle slope greater than 1.65 is only 0.38%.  
In hypothesis testing, the significance level is commonly set at 0.05 (5%), meaning that if the probability of obtaining the observed result is below 5%, the null hypothesis is rejected. Since the p-value obtained in this study is smaller than the 5%, sufficient evidence exists to reject the null hypothesis.    
This result provides statistical support for the alternative hypothesis, suggesting that the true slope may be greater than 1.5 and possibly exceed 1.65.


# Understanding the Importance of Bias in Big Data

1. *The Hidden Biases in Big Data*
* The Twitter data from Hurricane Sandy gave a misleading picture of the disaster. Most tweets came from Manhattan, but the worst damage happened in places like Breezy Point, Coney Island, and Rockaway. These areas had fewer tweets because of power outages and no internet access. As a result, the data made it seem like Manhattan was the center of the disaster, even though it was not.
* Boston’s StreetBump app uses smartphone sensors to detect potholes. However, not everyone owns a smartphone, especially low-income or older people. This means some areas did not report potholes, even if they had many. As a result, the city might have focused repairs on areas with more reports, ignoring other places that needed help.
* A similar issue can arise in building energy management due to cognitive biases in decision-making. According to Rakitta and Wernery (2021), people do not always make rational choices when selecting energy-efficient technologies. Their study found that biases such as framing, anchoring, default, and decoy effects can significantly influence decisions​
. For example, if a building contractor presents a less energy-efficient home as the default option, many people will accept it without actively considering better alternatives. This happens because people tend to stick with the given choice rather than explore other options. As a result, even if a more sustainable design is available, it may not be widely adopted simply due to how the options are presented.  
This issue is closely related to biases in big data. If energy policies or research studies rely solely on data collected from biased decision-making processes, they may misrepresent actual energy usage and consumer preferences. For example, a dataset showing that most people choose less efficient buildings might falsely suggest that people do not care about sustainability, when in reality, they were just influenced by the way choices were framed. This kind of bias can lead to misallocation of resources and ineffective energy policies, reinforcing existing inefficiencies instead of promoting optimal solutions. To address this, decision-makers must consider human behavior alongside big data analytics, ensuring that research and policies support fair and sustainable energy solutions.

2. *Biases in machine learning models and big data analytics: The international criminal and humanitarian law implications*
* The data used for machine learning models and big data analytics often contain biases, which can lead to unfair or incorrect results. Some common issues include selection bias, where certain groups or events are over- or under-represented; reporting bias, where data from some sources is more available than others; and implicit bias, where models reflect human prejudices. For example, if an ML model is trained using social media data, it may favor information from certain groups while ignoring others, leading to misleading conclusions about real-world events.
* To reduce bias in machine learning models, data augmentation and balancing techniques can be applied. One effective approach is counterfactual data augmentation, which involves modifying existing data to create alternative scenarios. For example, researchers have used gender-swapping techniques in text datasets to prevent models from reinforcing gender stereotypes (Zmigrod et al., 2019). Additionally, reweighting training samples can help balance underrepresented groups by adjusting their influence on the model’s learning process.  
Another method is bias auditing, where fairness-checking tools—such as IBM’s AI Fairness 360 (AIF360)—are used to detect and mitigate biases before deployment. AIF360 is an open-source Python toolkit that provides various algorithms and metrics to evaluate and improve fairness in machine learning models (IBM, 2021). By carefully curating diverse datasets and applying bias correction techniques, machine learning models can become more equitable and reliable.

3. *Big data in Earth science: Emerging practice and promise*
* Among the three major areas discussed in the paper, I chose Global Climate Models (GCMs), which use big data to simulate Earth’s climate by incorporating physical, chemical, and biological processes. The Coupled Model Intercomparison Project (CMIP6) enhances climate predictions by integrating high-resolution data, improving the accuracy of climate processes such as ocean-atmosphere interactions and severe weather forecasting. However, biases and missing data can affect results. These challenges highlight the need for ongoing refinements in data collection and model validation to improve climate predictions​.
